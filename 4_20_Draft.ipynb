```python
files = glob.glob("data/*fast.csv")
from datetime import datetime as dt
# IF you run into an error in this cell, ask xtian
# there's a bug in the data
# BAD_F = "data/b827ebbe2801_1521218535.02_fast.csv"
df = pd.concat(pd.read_csv(f) for f in files)
```

```python
df420 = df
df420["timestamp"] = df420["time"].apply(dt.fromtimestamp)
df420 = df420.sort_values("timestamp")
```

```python
df420 = df420[['time', 'dBAS', 'timestamp']]
print (len(df420))
df420.head()
```

```python
# make dataframe only while it was recording outside
# '2018-04-20 18:48:00.000' -ish
df420 = df420[(df420.timestamp > '2018-04-20 00:00:00.000') & (df420.timestamp < '2018-04-20 18:48:00.000')]
print (len(df420))
df420.head()
```

```python
df420[(df420.timestamp > '2018-04-20 00:00:00.000') & (df420.timestamp < '2018-04-20 18:48:00.000')].set_index("timestamp" )['dBAS'].plot(figsize=(16,9))
```

```python
fig = plt.figure(figsize=(30,18))
ax = fig.add_subplot(111)

# create rolling mean
noise_rmean = pd.rolling_mean(df420.dBAS, 1000, center=False)

# rough guess to play with vis, cutting out last weird chunk
ax.plot(df420.time, df420.dBAS, '--', label='dBAS - noise', alpha=.1)
ax.plot(df420.time, noise_rmean, linewidth=3, label="Noise Rolling Mean")

plt.title('Gowanus Noise - April 20, 2018', fontsize=40)
plt.ylabel('dBAS', fontsize=30)
plt.xlabel('Time', fontsize=30)
plt.legend(prop={'size': 23}, frameon=True)
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.grid(linewidth=.2)
```

```python
# bringing in air quality 
df_air = pd.read_csv('data/420_AQ.log', skiprows=1, header=None, names=["PM25", "PM10"])
print (len(df_air))
df_air.head()
```

```python
from datetime import date, datetime, timedelta

def perdelta(start, end, delta):
    time = []
    curr = start
    while curr < end:
        time.append(curr)
        curr += delta
    return time
    
# pd.DataFrame(perdelta(datetime(2018,4,20,10,30),datetime(2018,4,20,14,2), timedelta(minutes=1)))    
```

```python
# adding on timestamp column at one min interval
df_air['timestamp'] = pd.DataFrame(perdelta(datetime(2018,4,20,10,30),datetime(2018,4,20,14,2), timedelta(minutes=1)))    
```
